Q3: loss function


```python
import numpy as np

def L1(yhat, y):
    loss = np.sum(np.abs(y - yhat))
    return loss

yhat = np.array([.9, 0.2, .1, .4, .9])
y = np.array([1, 0, 0, 1, 1])
l1_loss = L1(yhat, y)
print("L1 = " + str(l1_loss))
```

```python
def L2(yhat, y):
    loss = np.sum((y - yhat) ** 2)
    return loss

yhat = np.array([.9, 0.2, .1, .4, .9])
y = np.array([1, 0, 0, 1, 1])
l2_loss = L2(yhat, y)
print("L2 = " + str(l2_loss))
```

```python
#using for loop
import numpy as np
import time

t1 = time.time()
def slow_dot_product(a, b):
    result = 0
    for i in range(len(a)):
        result += a[i] * b[i]
    return result

a = np.random.randn(1_000_000)
b = np.random.randn(1_000_000)
result = slow_dot_product(a, b)
t2 = time.time()
t2 - t1
```
