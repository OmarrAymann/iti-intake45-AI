3. Data Preparation: Cleaning and Formatting

```python
#large number of missing values in this columns
per_of_nan = df.isna().sum() / len(df) * 100
percent_of_nan = per_of_nan[per_of_nan > 20]
percent_of_nan.sort_values(ascending=False).plot.bar(title='percentage of NaN > 25%', figsize=(10, 5))
plt.ylabel('percent of NaN')
plt.show()
```

```python
#drop columns with more than 25% missing values 
df.drop(columns=percent_of_nan.index, inplace=True)
df.shape
```

```python
# unnecessary columns
droped_columns=['ID number','year','ZIP code','Longitude location of institution','Latitude location of institution','Religious affiliation']
df.drop(columns=droped_columns ,inplace=True)
df.shape
```
